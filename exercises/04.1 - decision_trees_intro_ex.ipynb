## Funkcja rysująca sektory decyzyjne

```python
def plot_decision_boundary(clf, X, y, title):
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                         np.linspace(y_min, y_max, 300))

    Z = np.array([clf(x) for x in np.c_[xx.ravel(), yy.ravel()]])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k')
    plt.title(title)
    plt.show()
```

## Podpunkt e) - Spirala

### Generacja danych

```python
def make_spiral(n=500, noise=0.01):
    n_half = n // 2
    theta = np.linspace(0, 4*np.pi, n_half)

    r = np.linspace(0.1, 0.5, n_half)
    x1 = 0.5 + r * np.cos(theta) + noise * np.random.randn(n_half)
    y1 = 0.5 + r * np.sin(theta) + noise * np.random.randn(n_half)

    x2 = 0.5 + r * np.cos(theta + np.pi) + noise * np.random.randn(n_half)
    y2 = 0.5 + r * np.sin(theta + np.pi) + noise * np.random.randn(n_half)

    X = np.vstack([np.c_[x1, y1], np.c_[x2, y2]])
    y = np.array([0]*n_half + [1]*n_half)
    return X, y

X, y = make_spiral()

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Zadanie 1 e)")
plt.show()
```
