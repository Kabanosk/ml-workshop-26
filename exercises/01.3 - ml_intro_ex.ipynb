# Tajemnice Ogrodu Botanicznego we Wrocawiu 
Jeste botanikem badajcym irysy w Ogrodzie Botanicznym Uniwersytetu Wrocawskiego! Chcesz przeanalizowa r贸偶ne gatunki tych kwiat贸w na podstawie ich cech, takich jak dugo i szeroko patk贸w. Wykorzystasz do tego prawdziwy zbi贸r danych 'Iris Dataset'.

```python
# Pobieranie danych
data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ["Dugo dziaki kielicha", "Szeroko dziaki kielicha", "Dugo patka", "Szeroko patka", "Gatunek"]
df = pd.read_csv(data_url, names=columns)
df
```

Uniwersytet chce 偶eby sprawdzi czy dugo patka i jego szeroko s jako powizane. Ty jako super mega dobry specjalista od uczenia maszynowego m贸gby zrobi to w ten spos贸b:

```python
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.scatter(df['Dugo patka'], df['Szeroko patka'])
plt.xlabel("Dugo patka")
plt.ylabel("Szeroko patka")
plt.title("Zale偶no dugoci do szerokoci patk贸w")

# Wykresy po prawej - kolorowe
plt.subplot(1, 2, 2)
colors = {species: color for species, color in zip(df['Gatunek'].unique(), ['red', 'blue', 'green'])}
for species in df['Gatunek'].unique():
    subset = df[df['Gatunek'] == species]
    plt.scatter(subset['Dugo patka'], subset['Szeroko patka'], label=species, color=colors[species])
plt.xlabel("Dugo patka")
plt.ylabel("Szeroko patka")
plt.title("Zale偶no dugoci do szerokoci patk贸w (z gatunkami)")
plt.legend(title='Gatunek')
plt.show()
```
