### Listy (Twój ekwipunek)

Gdy chcesz trzymać wiele rzeczy w jednym miejscu. Pamiętaj: w programowaniu liczymy od zera!

```python
ekwipunek = ["Miecz", "Łuk", "Mikstura", "Złoto", "Mapa"]
```

### Wyciąganie rzeczy z listy

Wyciągamy pierwszy i ostatni przedmiot:


```python
pierwszy_item = ekwipunek[0]   # "Miecz"
ostatni_item = ekwipunek[-1]   # "Mapa" (minus jeden to zawsze koniec, a -2 to drugi od końca)
```

### Slicing (Wycinanie kawałków listy)

Co jeśli chcesz wyciągnąć np. 3 pierwsze przedmioty naraz? Używamy magicznego dwukropka `:`. Działa to jak zaznaczanie tekstu myszką.

```python
# Od indeksu 0 do 3 (ale BEZ trójki! Czyli weźmie 0, 1 i 2)
poczatek = ekwipunek[0:3]      # ["Miecz", "Łuk", "Mikstura"]
print("Początek", poczatek)

# Jeśli zaczynamy od zera, możemy je pominąć
szybki_poczatek = ekwipunek[:3]
print(f"Jak widać pierwsza opcja - [0:3] ({poczatek}) i druga opcja - [:3] ({szybki_poczatek}), są takie same!")

# Od indeksu 2 aż do samego końca (pamiętamy że liczymy od 0)
koncowka = ekwipunek[2:]
print("Koncówka", koncowka)
```

### Magiczny trik ze Slicingiem (Odwracanie)

Chcesz przeczytać listę od tyłu? Wystarczy użyć dwóch dwukropków i minus jedynki.

```python
odwrocony_ekwipunek = ekwipunek[::-1]
print(f"Normalny ekwipunek={ekwipunek}")
print(f"Odwrócony ekwipunek={odwrocony_ekwipunek}")
```

### Instrukcje warunkowe If/Else

Logika gry. Zrób coś TYLKO jeśli warunek jest spełniony.

```python
if hp <= 0:    # jeśli straciłeś całe życie - umierasz
    print("Game Over! ☠️☠️☠️")
elif hp < 20:  # jeśli masz hp > 0, ale mniejsze niż 20 to wykonujemy kod poniżej
    print("Uwaga, mało życia! ⚠️")
else:          # jak żaden warunek nie jest spełniony wykonaj kod poniżej
    print("Walcz dalej! ⚔️")
```

### Pętla FOR

Nie chce Ci się pisać tego samego 100 razy? Użyj pętli. Kod wewnątrz wykona się dla każdego elementu z listy.

```python
for przedmiot in ekwipunek:
    print(f"Znalazłeś: {przedmiot}")
```

### Pętla FOR z licznikiem

Moglibyśmy chcieć zrobić `id` dla każdego przedmiotu.

Pamiętasz - `ekwipunek[i]` to element z listy ekwipunek o numerze i (czyli na początku 0, potem 1, potem 2. Jakbyśmy mięli dłuższą listę to 3, 4, 5...)

```python
for i in range(len(ekwipunek)):
  print(f"Przedmiot o id={i} nazywa się: {ekwipunek[i]}")
```

### Funkcje (Tworzenie własnych zaklęć)

Zamykasz kawałek kodu w paczkę, nadajesz jej nazwę i wywołujesz kiedy tylko chcesz.

```python
def lecz_postac(aktualne_hp, mikstura_leczenia):
    nowe_hp = aktualne_hp + mikstura_leczenia
    return nowe_hp
```

### Używanie własnej funkcji

Żeby użyć funkcji musimy włożyć do niej argumenty, robimy to za pomocą okrągłych nawiasów. Poniżej do funkcji `lecz_postac` podajemy argumenty `aktualne_hp` i `mikstura_leczenia`, gdzie jako `aktualne_hp` przyjmiemy wartość zmiennej `hp` (czyli 100), a jako `mikstura_leczenia` (o ile uleczymy naszego bohatera) przyjmiemy 50 (bo tyle podaliśmy).

```python
hp = lecz_postac(hp, 50)
print(f"Moje aktualne hp: {hp}")
```

Zwykłe listy w Pythonie są fajne, ale wyobraź sobie, że musisz policzyć średnią ze 100 tysięcy ocen z całej Polski. Zwykły Python by się spocił i zawiesił. Numpy to takie matematyczne Ferrari - robi obliczenia na ogromnych zbiorach danych w ułamku sekundy!



### Jak zaimportować?

```python
import numpy as np
```

### Tworzenie "Array" (super-tablicy)

Array to podstawowy klocek, z którego budujemy nasze dane do AI. Wygląda trochę jak lista, którą już poznaliśmy wcześniej!

```python
zwykla_lista = [1, 2, 3, 4, 5]
super_tablica = np.array(zwykla_lista)
print(super_tablica)
```

### Średnia

Wyobraź sobie, że to czasy przejścia poziomu w grze. Ile średnio zajmuje ten poziom?

```python
czasy_w_grze = np.array([12, 15, 10, 20, 11])
srednia = np.mean(czasy_w_grze)
print(srednia)
```

### Mediana

Wartość "środkowa". Lepsza niż średnia, gdy np. jeden gracz poszedł na obiad w trakcie gry i zawyżył wynik.

```python
czasy_w_grze = np.array([12, 15, 10, 200, 11]) # Ten z 200 to ten od obiadu!
mediana = np.median(czasy_w_grze)
print(mediana)
```

### Wartość minimalna i maksymalna

Kto był najszybszy, a kto najwolniejszy?

```python
czasy_w_grze = np.array([12, 15, 10, 20, 11])
najszybszy = np.min(czasy_w_grze)
najwolniejszy = np.max(czasy_w_grze)
print("najszybszy:", najszybszy)
print("najwolniejszy:", najwolniejszy)
```

## 🤖 Scikit-Learn (sklearn) - Fabryka Sztucznej Inteligencji

W `scikit-learn` znajdują się wszystkie popularne modele uczenia maszynowego.

Pamiętaj: schemat działania to ZAWSZE: Wybierz model -> Naucz go (.fit) -> Zgaduj (.predict).

### Import i przygotowanie danych

Duże X to cechy (np. wiek i waga), a małe y to etykiety (np. 0=Kot, 1=Pies).

```python
from sklearn.neighbors import KNeighborsClassifier

X = np.array([[1, 2], [2, 3], [8, 9], [9, 10]])
y = np.array([0, 0, 1, 1])
```

### Wybierz i skonfiguruj model

Wybieramy naszą broń. Tutaj algorytm KNN (pytamy 3 najbliższych sąsiadów).

```python
model = KNeighborsClassifier(n_neighbors=3)
```

### Trenowanie modelu (FIT)

Magia! Tutaj model się uczy. Odkrywa zależności między cechami (X) a odpowiedziami (y).

```python
model.fit(X, y)
```

### Przewidywanie (PREDICT)

Przyszedł nowy, nieznany zwierzak. Zgadnij, co to jest!

```python
nowy_zwierzak = np.array([[7, 8]])
przewidywanie = model.predict(nowy_zwierzak)
print(przewidywanie)
```

### Ocenianie modelu (SCORE)

Wrzucamy mu testowe dane (kartkówkę) i patrzymy, jak dobrze sobie poradził (wynik od 0.0 do 1.0).

```python
X_test = np.array([[1, 2], [5, 9], [0,2]])
y_test = np.array([0, 1, 0])

dokladnosc = model.score(X_test, y_test)
dokladnosc
```
